```python
# 导入包
import pandas as pd 
import numpy as np
```

```python
# 读取数据集
ds1 = pd.read_csv('./dataset/01handing_sport_dataset.csv')
ds2 = pd.read_csv('./dataset/02handing_normal_dataset.csv')
ds3 = pd.read_csv('./dataset/03Double lane change_sport_dataset.csv')
ds4 = pd.read_csv('./dataset/04Double lane change_normal_dataset.csv')
ds5 = pd.read_csv('./dataset/05sine-sweep_normal_dataset.csv')
ds6 = pd.read_csv('./dataset/06sine-sweep_sport_dataset.csv')
ds7 = pd.read_csv('./dataset/07sine steer_sport_dataset.csv')
ds8 = pd.read_csv('./dataset/08sine steer_normal_dataset.csv')
ds9 = pd.read_csv('./dataset/09step steer_sport_dataset.csv')
ds10 = pd.read_csv('./dataset/10step steer_normal_dataset.csv')
ds11 = pd.read_csv('./dataset/11steady state corner_sport_dataset.csv')
ds12 = pd.read_csv('./dataset/12steady state corner_normal_dataset.csv')
ds13 = pd.read_csv('./dataset/13Acceleration barking_sport_dataset.csv')
ds14 = pd.read_csv('./dataset/14Acceleration barking_normal_dataset.csv')
```

## 将各个工况的数据集叠加起来

```python
dataset = pd.concat([ds1, ds2, ds3, ds4, ds5, ds6, ds7, ds8, ds9, ds10, ds11, ds12, ds13, ds14], axis=0, ignore_index=True)
```
